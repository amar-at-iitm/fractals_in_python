$$
I=[x_0,x_N]\subset \mathbb{R}, \qquad x_0<x_1<\cdots <x_N .
$$

Let

$$
f\in C(I)
$$

be the original function, and let

$$
g\in C(I)
$$

be a base function satisfying the endpoint interpolation conditions

$$
g(x_0)=f(x_0), \qquad g(x_N)=f(x_N).
$$

Let

$$
\Delta=\{x_0,x_1,\ldots,x_N\}
$$

be a partition of $I$. For each $i=1,2,\ldots,N$, define an affine map

$$
L_i:I\to I_i=[x_{i-1},x_i]
$$

by

$$
L_i(x) = a_i x+b_i,
$$

where

$$
a_i=\frac{x_i-x_{i-1}}{x_N-x_0},
$$

and

$$
b_i=\frac{x_Nx_{i-1}-x_0x_i}{x_N-x_0}.
$$

Thus,

$$
L_i(x_0)=x_{i-1}, \qquad L_i(x_N)=x_i.
$$

The inverse map is

$$
L_i^{-1}(x)=\frac{x-b_i}{a_i}, \qquad x\in [x_{i-1},x_i].
$$

Equivalently,

$$
L_i^{-1}(x) = x_0 + \frac{(x-x_{i-1})(x_N-x_0)}{x_i-x_{i-1}}.
$$

Let

$$
\alpha=(\alpha_1,\alpha_2,\ldots,\alpha_N)
$$

be a scale vector satisfying

$$
|\alpha_i|<1, \qquad i=1,2,\ldots,N.
$$

The $\alpha$-fractal function associated with $f$, the base function $g$, the partition $\Delta$, and the scale vector $\alpha$ is the unique continuous function

$$
f^\alpha:I\to \mathbb{R}
$$

satisfying the self-referential equation

$$
f^\alpha(x)=f(x)+\alpha_i\left[f^\alpha\left(L_i^{-1}(x)\right)-g\left(L_i^{-1}(x)\right)\right], \qquad x\in [x_{i-1},x_i],
$$

for every

$$
i=1,2,\ldots,N.
$$

Equivalently,

$$
f^\alpha(L_i(x))=f(L_i(x))+\alpha_i\left[f^\alpha(x)-g(x)\right],\qquad x\in I.
$$

The above equation can also be written as

$$
f^\alpha=f+\alpha_i\left(f^\alpha-g\right)\circ L_i^{-1} \quad \text{on }[x_{i-1},x_i].
$$

Define the operator

$$
T:C(I)\to C(I)
$$

by

$$
(T\varphi)(x) = f(x) + \alpha_i\left[\varphi\left(L_i^{-1}(x)\right) - g\left(L_i^{-1}(x)\right)\right], \qquad x\in [x_{i-1},x_i].
$$

Then $f^\alpha$ is the unique fixed point of $T$, that is,

$$
T f^\alpha=f^\alpha.
$$

Moreover, if

$$
\|\alpha\|_\infty
=
\max_{1\le i\le N}|\alpha_i|<1,
$$

then $T$ is a contraction on $C(I)$ with respect to the uniform norm

$$
\|\varphi\|_\infty
=
\sup_{x\in I}|\varphi(x)|.
$$

Indeed, for $\varphi,\psi\in C(I)$,

$$
\|T\varphi-T\psi\|_\infty
\le
\|\alpha\|_\infty
\|\varphi-\psi\|_\infty .
$$

Therefore, by the Banach fixed point theorem, $T$ has a unique fixed point, which is the $\alpha$-fractal function $f^\alpha$.

The fractal perturbation can be written as

$$
f^\alpha-f
=
\alpha_i
\left[
f^\alpha-g
\right]\circ L_i^{-1}
\quad
\text{on }[x_{i-1},x_i].
$$

Taking the uniform norm gives the estimate

$$
\|f^\alpha-f\|_\infty
\le
\|\alpha\|_\infty
\|f^\alpha-g\|_\infty.
$$

Since

$$
\|f^\alpha-g\|_\infty
\le
\|f^\alpha-f\|_\infty+\|f-g\|_\infty,
$$

we obtain

$$
\|f^\alpha-f\|_\infty
\le
\|\alpha\|_\infty
\left(
\|f^\alpha-f\|_\infty+\|f-g\|_\infty
\right).
$$

Hence,

$$
\left(1-\|\alpha\|_\infty\right)
\|f^\alpha-f\|_\infty
\le
\|\alpha\|_\infty
\|f-g\|_\infty.
$$

Therefore,

$$
\boxed{
\|f^\alpha-f\|_\infty
\le
\frac{\|\alpha\|_\infty}{1-\|\alpha\|_\infty}
\|f-g\|_\infty
}
$$

provided

$$
\|\alpha\|_\infty<1.
$$

In particular, when

$$
\alpha=0,
$$

we recover the original function:

$$
f^0=f.
$$

$$
\left(f^\alpha(x)\right)''=
f''(x)+\alpha_i\left[
\left(f^\alpha\right)''\!\left(L_i^{-1}(x)\right)\left(\frac{d}{dx}L_i^{-1}(x)\right)^2
+\left(f^\alpha\right)'\!\left(L_i^{-1}(x)\right)\frac{d^2}{dx^2}L_i^{-1}(x)
-g''\!\left(L_i^{-1}(x)\right)\left(\frac{d}{dx}L_i^{-1}(x)\right)^2
-g'\!\left(L_i^{-1}(x)\right)\frac{d^2}{dx^2}L_i^{-1}(x)
\right]
$$

Since $L_i$ is usually affine, $L_i^{-1}(x)$ is also affine, so $\dfrac{d^2}{dx^2}L_i^{-1}(x)=0$. Then this simplifies to

$$
\left(f^\alpha(x)\right)''=
f''(x)+\alpha_i\left[
\left(f^\alpha\right)''\!\left(L_i^{-1}(x)\right)
-g''\!\left(L_i^{-1}(x)\right)
\right]
\left(\frac{d}{dx}L_i^{-1}(x)\right)^2
$$

$$
\left(f^\alpha(x)\right)'' = f''(x) + \alpha_i\frac{\left[\left(f^\alpha\right)''\!\left(L_i^{-1}(x)\right) - g''\!\left(L_i^{-1}(x)\right)\right]}{a_i^2}, \qquad x\in [x_{i-1},x_i].
$$

\left(f^\alpha(x)\right)'' = f''(x)+\alpha_i\left(\frac{x_N-x_0}{x_i-x_{i-1}}\right)^2\left[\left(f^\alpha\right)''\!\left(L_i^{-1}(x)\right) - g''\!\left(L_i^{-1}(x)\right)\right], \qquad x\in [x_{i-1},x_i].
